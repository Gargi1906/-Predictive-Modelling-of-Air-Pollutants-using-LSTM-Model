# **Other Method of Data Extraction**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, beta, weibull_min

# Load the dataset
df = pd.read_csv('2023-10-17 07-33-23.242543 - Week-1.csv')

# Extract temperature, humidity, and wind speed columns
temperature_data = df['Temperature'].dropna()
humidity_data = df['Humidity'].dropna()
wind_speed_data = df['Wind Speed (mph)'].dropna()


# Plot histograms for visual inspection
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(temperature_data, kde=True)
plt.title('Temperature Distribution')

plt.subplot(1, 3, 2)
sns.histplot(humidity_data, kde=True)
plt.title('Humidity Distribution')

plt.subplot(1, 3, 3)
sns.histplot(wind_speed_data, kde=True)
plt.title('Wind Speed Distribution')

plt.tight_layout()
plt.show()
```

```python
# Function to fit distribution and plot
def fit_and_plot(data, distribution, color, label):
    params = distribution.fit(data)
    x = np.linspace(data.min(), data.max(), 10)
    y = distribution.pdf(x, *params)
    plt.plot(x, y, color, label=label)
```

```python
plt.figure(figsize=(15, 5))

fit_and_plot(temperature_data, norm, 'b--', 'Temperature Fit (Normal)')
fit_and_plot(humidity_data, beta, 'g--', 'Humidity Fit (Beta)')
fit_and_plot(wind_speed_data, weibull_min, 'orange', 'Wind Speed Fit (Weibull)')

# Plot the original data
sns.histplot(temperature_data, kde=True, color='blue', alpha=0.5, label='Temperature Data')
sns.histplot(humidity_data, kde=True, color='green', alpha=0.5, label='Humidity Data')
sns.histplot(wind_speed_data, kde=True, color='orange', alpha=0.5, label='Wind Speed Data')

plt.legend()
plt.show()
```
